```python
import pandas as pd

df = pd.read_csv('2019.csv')
```

```python
df.head()          # Show the first few rows
df.info()          # Overview of data types and null values
df.describe()      # Summary statistics
```

```python
import matplotlib.pyplot as plt

df['Overall rank'].hist(bins=20)
plt.xlabel('Overall rank')
plt.ylabel('GDP per capita')
plt.title('Distribution of Column')
plt.show()
```

```python
print(df.describe())
print(df.duplicated().sum())  # Check for duplicate rows
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = ['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 
                     'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']

for col in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
sns.scatterplot(x='GDP per capita', y='Score', data=df)
plt.title('Happiness Score vs GDP per capita')
plt.show()
```

```python
sns.scatterplot(x='Social support', y='Score', data=df)
plt.title('Happiness Score vs Social Support')
plt.show()
```
